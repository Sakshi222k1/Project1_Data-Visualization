```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("time_series_2019-ncov-Confirmed.csv")
df
```

```python
d=df.describe()
dates=d.columns
dates
```

```python
drop_column=list()
for i in range(len(dates)):
    count=0
    for j in range(487):
        if (df.iloc[j][i+4]==0):
            count+=1
    per=count/487
    if (per>=0.71):
        drop_column.append(dates[i])
drop_column
```

```python
df=pd.read_csv('time_series_2019-ncov-Confirmed.csv')
df
for i in drop_column:
    df=df.drop(i,axis=1)
df
```

```python
df1=df.groupby('Country/Region').sum()
df1
```

```python
cases=[]
countries=[]
for j in range(171):
        v=df1.iloc[j][-1]
        cases.append(v)
c1=df1.index
for i in c1:
    countries.append(i)
    c=0
for a in range(len(countries)):
    for b in range(len(cases)):
        if a==b:
            print("Country:",countries[a],end="")
            print("Cases:",cases[b])
            c=c+1
print("Total no. of affected contries:",c)
```

```python
plt.figure(figsize=(20,30))
plt.title("PIE CHART SHOWING COVID CASES IN DIFFERENT COUNTRIES",size=30)
plt.pie(cases)
plt.legend(countries,loc="best",prop={'size':20})
plt.show()
```

```python
dates1=df1.keys()
dates1
```

```python
d_cases=[]
for i in range(len(dates1)):
    s=df1[dates1[i]].sum()
    d_cases.append(s)
print(d_cases)
for a in range(len(dates1)):
    for b in range(len(d_cases)):
        if a==b:
            print("Date:",dates1[a],end="")
            print("Cases:",d_cases[b])
```

```python
plt.figure(figsize=(20,30))
plt.title("PIE CHART SHOWING COVID CASES ON DIFFERENT DAYS",size=30)
plt.pie(d_cases)
plt.legend(dates,loc="best",prop={'size':20})
plt.show()
```
